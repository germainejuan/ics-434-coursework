#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

discipline = ['CS', 'Engineering', 'Math', 'Physics'] # Create an arbitrary array-like instead of integers
np.random.seed(42) # Set 42 as seed
students_affiliations = np.random.choice(discipline, size=1000) # Generate a random sample from discipline with a size of 1000


students_affiliations_df = pd.DataFrame(data=students_affiliations, columns=['Counts']) # Create a dataframe from the random sample
students_affiliations_value = students_affiliations_df['Counts'].value_counts() # Creates count per discipline
students_affiliations_value = students_affiliations_value.to_frame() # Aligns 'Counts' column with its values
students_affiliations_value = students_affiliations_value.rename_axis('Disciplines') # Renames column to 'Disciplines'
students_affiliations_value = students_affiliations_value.reset_index() # Creates Index

# Assign columns into variables
Disciplines = students_affiliations_value['Disciplines'].to_numpy()
Counts = students_affiliations_value['Counts'].to_numpy()

plt.bar(Disciplines, Counts , color='#FF5733', alpha=0.55) # Display bar graph
plt.title('Distribution of Students in the Program') # Add graph title
plt.xlabel('Disciplines') # Add x axis
plt.ylabel('Counts') # A y axis
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(2092022) # Set seed as 2092022

N = 100 # Number of values
x = np.random.rand(N) # Generate N random x values
y = np.random.rand(N) # Generate N random y values

# Create scatter plot
# marker "," changes the symbol to squares
plt.scatter(x, y, marker=',')
```
